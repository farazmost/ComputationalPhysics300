Now we try to build the One Rule Classification model for learnign to our data using the instructions of:
The algorithm is the following. 

- 1, For each predictor,
     - 1.1 For each value of that predictor, make a rule as follows;
           - 1.1.1 Count how often each value of target (class) appears
           - 1.1.2 Find the most frequent class
           - 1.1.3 Make the rule assign that class to this value of the predictor
     - 1.2 Calculate the total error of the rules of each predictor
- 2, Choose the predictor with the smallest total error.




```python
def OneRuleFrequency(x, Group):
    """
    The input for the function is an array containing values of only one of the columns, therefore:
    Sepal Length, Sepal Width, Petal Length, Petal Width
    """
    zerocounter=0 #Create a counter for the number of zeros and ones we have for our frequency
    onecounter=0
    zeros=[] #Create a list of those values that have zeros or ones
    ones=[]
    y=Y
    
    
    
    for i in range(len(x)):
        if x[i] == 0:
            zerocounter += 1
            zeros.append(i)
        else:
            onecounter += 1
            ones.append(i)
            
    flowertype1, flowertype2, flowertype3 = 0, 0, 0
    
    for j in ones:
        if y[j]== 0:
            flowertype1 += 1
        elif y[j]== 1:
            flowertype2 += 1
        elif y[j]== 2:
            flowertype3 += 1
    zeroflowertype1 = 50 - flowertype1
    zeroflowertype2 = 50 - flowertype2 
    zeroflowertype3 = 50 - flowertype3
    data_d = OrderedDict({str(Group): [1,0],
                          'Setosa': [flowertype1,zeroflowertype1],
                          'Veriscolour':[flowertype2,zeroflowertype2],
                          'Virginica':[flowertype3,zeroflowertype3]}) 
    data_table = pd.DataFrame(data_d, index= ['',''])
    
    return data_table
    
Group = ['sepal length', 'sepal width', 'petal length', 'petal width']
for i in range(len(Group)):
    print(OneRuleFrequency(X_d[:,i],Group[i]),'\n')
    
```

After going through this, I realized that using a binary classification of our original data doesnt make sense, due to there being 3 outcomes, but with 2 classification rules, therefore, with the help of Stanley Baronett's Code, We are able to classify the data into 3 bins instead of the binary (bin=2):

```python
"""
Code taken from Stanley Baronett. This is the code to distribute our data into the 3 bins:
EDGES are the boundaries of our bins and the limits of where the data can fall into
"""
def rule(a):
    """
    Returns rule values based on optimal bin edges derived
    from a species' mean value for a given attribute.
    
    Parameters
    ----------
    a : array_like
        Array containing attribute values grouped by species
    
    Returns
    -------
    edges : array_like
        Array containing rule values based on bin edges
    """
    means = [] # hold attribute means for each species
    means.append(np.mean(a[:50]))     # setosa
    means.append(np.mean(a[50:100]))  # versicolor
    means.append(np.mean(a[100:150])) # virginica
    # take the average value between means to be edges
    edges = [(means[0] + (means[1] - means[0])/2),
             (means[1] + (means[2] - means[1])/2)]
    # correct for cases where edges are decreasing
    if edges[0] > edges[1]:
        temp = edges[0]
        edges[0] = edges[1]
        edges[1] = temp
    return edges

```

```python
"""
Now we can define the bounds of our bins for each of groups
"""
for i in range(4):
    boundaries = rule(X[:,i])
    boundaries.insert(0,np.min(X[:,i]))
    boundaries.append(np.max(X[:,i]))
    print ("Boundaries for", Group[i],"is", boundaries,'\n')
print("So for example, the Petal Width bounds for the bins are:",'\n' 
      "[","%.2f" %boundaries[0],",","%.2f" %boundaries[1],"]",'\n'
      "[","%.2f" %boundaries[1],",","%.2f" %boundaries[2],"]",'\n'
      "[","%.2f" %boundaries[2],",","%.2f" %boundaries[3],"]",'\n'
     )
```

###  So now we have to find the frequency of distribution within these bins:

```python
"""
Code from Stanley Baronett
"""
def freqs(a, rule):
    bins = np.zeros([3,3]) # hold bin counts
    counts = np.zeros(3)
    start, end = 0, 50

    for i in range(3): # for each species
        counts[0] = np.where(a[start:end] < rule[0])[0].size
        counts[2] = np.where(a[start:end] >= rule[1])[0].size
        counts[1] = np.where(a[start:end] >= rule[0])[0].size - counts[2]
        bins[i,:] = counts
        start += 50
        end += 50
        
    return bins

def freq_table(a, rule, prop):
    labels = ['(0, '+f'{rule[0]:.2f}'+')',
              '['+f'{rule[0]:.2f}'+', '+f'{rule[1]:.2f}'+')',
              '['+f'{rule[1]:.2f}'+', max.)']

    data = OrderedDict({prop: labels,
                        'Setosa': a[:,0],
                        'Versicolor': a[:,1],
                        'Virginica': a[:,2],})
    pd.options.display.float_format = '{:,.0f}'.format
    return pd.DataFrame(data)
```

```python
sl_rule = rule(X[:,0])
sl_bins = freqs(X[:,0], sl_rule)
freq_table(sl_bins, sl_rule, Group[0])
```

```python
sw_rule = rule(X[:,1])
sw_bins = freqs(X[:,1], sw_rule)
freq_table(sw_bins, sw_rule, Group[1])
```

```python
pl_rule = rule(X[:,2])
pl_bins = freqs(X[:,2], pl_rule)
freq_table(pl_bins, pl_rule, Group[2])
```

```python
pw_rule = rule(X[:,3])
pw_bins = freqs(X[:,3], pw_rule)
freq_table(pw_bins, pw_rule, Group[3])
```

```python
"""
Code from Stanley Baronett
"""
def accuracy(a):
    return np.amax(a, axis=1).sum() / a.sum()

results = np.asarray([accuracy(sl_bins),
                      accuracy(sw_bins),
                      accuracy(pl_bins),
                      accuracy(pw_bins)])
accuracies = OrderedDict({'Predictor': Group,
                            'Accuracy': results*100,})
pd.options.display.float_format = '{:,.2f} %'.format
pd.DataFrame(accuracies)
```
