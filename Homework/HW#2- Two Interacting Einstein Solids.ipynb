```python
import numpy as np
#import matplotlib as plt
import matplotlib.pyplot as plt
from math import log, exp

def factorial (N):
    N = int(N)
    total = 1
    if N < 0:
        print("Factorial error: N must be greater than 0")
    elif N > 1:
        for i in range(2, N+1):
            total *= i
    return total

    
def Multiplicity(N,n):
    return exp(log(factorial(N+n-1)) - log(factorial(n)) - log(factorial(N-1))) 

def Plot_Omega(Na,Nb,Q):
    qa_list = range (Q+1)
    Omega_a = []
    Omega_b = []
    #Omega_t = np.empty(1)
    for qa in qa_list:
        Omega_a.append(Multiplicity(Na,qa))
        Omega_b.append(Multiplicity(Nb,Q-qa))
        
    Omega_a = np.asarray(Omega_a)
    Omega_b = np.asarray(Omega_b)
    Omega_t = Omega_a * Omega_b
    
    plt.plot(qa_list, Omega_t, linewidth=2)
    plt.plot(qa_list, Omega_a, '-r')
    plt.plot(qa_list, Omega_b, '-g')
    plt.title('NA=' + str(Na) + ', NB=' + str(Nb) + ', q =' + str(Q)) 
    plt.xlabel('qa')
    plt.ylabel('Total Omega')
    plt.show()
    
```

```python
Plot_Omega(3, 2,100)
```

```python
Plot_Omega(300, 200,100)
```

```python
Plot_Omega(3000, 2000,100)
```
