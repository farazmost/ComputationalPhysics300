```python
from matplotlib import pyplot as plt
import numpy as np
def MandelBrot(N):
    #Define empty array
    M = np.zeros([N, N], float)
    #Define plot points in the imaginary coordinate system
    xvalues = np.linspace(-2, 2, N)
    yvalues = np.linspace(-2, 2, N)
    #testing all plot points
    for i in range(len(xvalues)): 
        for j in range(len(yvalues)):
            z = 0        
            c = xvalues[i] + yvalues[j]*1j       
            #iterations for (x,y) pairs
            for k in range(100):
                z = z*z + c
                if np.absolute(z) > 2:
                    M[j,i] = k
                    break
    #show the graph                
    plt.imshow(M,cmap='hot',extent=[-2,2,-2,2])
    plt.title('Mandelbrot Visualization')
    plt.colorbar()
    plt.show()
 
        
MandelBrot(1000)       
```
