```python
%matplotlib inline  
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=[12, 8])

f = lambda x: (np.sin(2*x))
a, b = 0,10
points = 100
order = 11

xset = np.linspace(a,b,points)
yset = f(xset) + np.random.rand(points) - 0.5

#do the curve fit
fit = np.polyfit(xset, yset, order, full=True)
y_p = np.polyval(fit[0], xset)


plt.plot(xset,f(xset), label='ideal function')
plt.plot(xset,yset, 'o', label='noisy data')
plt.plot(xset,y_p, label='polyfit'+str(order))


plt.xlabel('x')
plt.ylabel('$f(x)$')
plt.xlim([a,b])
plt.legend(fontsize=12)
plt.show()
```

Becuase sin(2x) is an odd function, I assumed a higher order odd number for our order
